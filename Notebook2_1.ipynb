# Controlled ODEs as features on the path space

In this notebook, we will investigate the effects of various control strategies on a controlled ODE (CODE). The system is controlled by different types of inputs, and we will solve the system both with and without control. By visualizing the results, we can observe how each type of control influences the system's behavior. Notably, CODE define (non-linear) features on the path space where the system lies, namely $\mathcal{C}^1([0,20], \mathbb{R}^2)$.

The system is described by the following controlled ODE:
$$ \dot{x}_1 = x_2 $$
$$ \dot{x}_2 = -x_1^3 + \cos(x_1) + u(t) $$

Where:
- $x_1$ is the position,
- $x_2$ is the velocity,
- \$u(t)$ is the control input.

We will explore five different control inputs and solve the system's equations numerically.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define a controlled nonlinear ODE
def controlled_ode(t, x, u_func):
    u = u_func(t)
    return np.array([x[1], -x[0]**3 + np.cos(x[0]) + u])
```
