# Introduction to TensorFlow and Keras

TensorFlow and Keras are powerful libraries for building and training deep learning models. In this notebook, we will train a simple feed-forward neural network using the two of them.

```python
# Installing Necessary Libraries

!pip install tensorflow matplotlib
```

```python
# Importing the necessary libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Generating random data
x = np.random.randn(100, 1).astype(np.float32)
y = x**2

plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Generated Data")
plt.show()
```

## **Building the Feed-Forward Neural Network**
We define a feed-forward neural network using Keras' Sequential API. The network consists of:
- An **input layer** with one neuron, since our input data consists of single values (`x`).
- Two **hidden layers**, each with 32 neurons and ReLU activation, which helps the network learn complex patterns.
- An **output layer** with one neuron, since we are predicting a single value (`y`).

```python
from tensorflow import keras
from tensorflow.keras import layers

# Defining the model
model = keras.Sequential([
    layers.Dense(32, activation='relu', input_shape=(1,)),  # First hidden layer
    layers.Dense(32, activation='relu'),                    # Second hidden layer
    layers.Dense(1)                                         # Output layer
])

# Compiling the model
# We use Adam optimizer for efficient learning.
# Mean Squared Error (MSE) is chosen as the loss function since we are dealing with regression.
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.01), loss='mse')
```

## **Training the Model**

We now train our model using the `.fit()` method.
- We pass in our training data (`x` and `y`).
- We specify `epochs=100`, meaning the model will iterate through the dataset 100 times.
- The `verbose=1` argument ensures we see progress updates during training.


```python
# Training the model
model.fit(x, y, epochs=100, verbose=1)
```

## **Making Predictions and Visualizing Results**

After training, we can use the model to make predictions and compare them with the actual values.
- We use `model.predict(x)` to generate predictions. Notice that in this simple example we have not separated a train and test set, as one would instead need to do in practical applications.
- We plot both the original dataset and the model's predictions to see how well the model learned the function `y = x^2`.

```python
# Predicting values
predictions = model.predict(x)
```

```python
# Plotting results
plt.scatter(x, y, label="Actual Data")
plt.scatter(x, predictions, label="Model Predictions", color='r', alpha=0.5)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Model Predictions vs Actual Data")
plt.show()
```
